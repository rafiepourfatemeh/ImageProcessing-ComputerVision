```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
square=np.zeros((10,10),np.uint8)
cv2.rectangle(square,(2,2),(7,7),2,2)
print(square)
```

```python
circle=np.zeros((10,10),np.uint8)
cv2.circle(circle,(5,5),4,3,-1)
print(circle)
```

```python
area1=np.zeros((300,300),np.uint8)
cv2.rectangle(area1,(50,50),(250,250),255,-1)
area2=np.zeros((300,300),np.uint8)
cv2.ellipse(area2,(150,150),(150,150),30,0,180,255,-1)
plt.subplot(121);plt.imshow(area1,cmap="gray");plt.title("square")
plt.subplot(122);plt.imshow(area2,cmap="gray");plt.title("ellipse")
```

```python
bitwise_and=cv2.bitwise_and(area1,area2)
bitwise_or=cv2.bitwise_or(area1,area2)
bitwise_not=cv2.bitwise_not(area1)
bitwise_xor=cv2.bitwise_xor(area1,area2)
plt.figure(figsize=[12,6])
plt.subplot(141);plt.imshow(bitwise_and,cmap="gray");plt.title("bitwise_and")
plt.subplot(142);plt.imshow(bitwise_or,cmap="gray");plt.title("bitwise_or")
plt.subplot(143);plt.imshow(bitwise_not,cmap="gray");plt.title("bitwise_not")
plt.subplot(144);plt.imshow(bitwise_xor,cmap="gray");plt.title("bitwise_xor")

```
