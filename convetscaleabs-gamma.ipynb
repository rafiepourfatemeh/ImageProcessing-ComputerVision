```python
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
image=cv2.imread(r"C:\\Users\Student\Desktop\pic02\006.jpg")
```

```python
m=np.ones(image.shape,dtype="uint8")*50
```

```python
added=cv2.add(image,m)
sub=cv2.subtract(image,m)
plt.figure(figsize=[15,5])
plt.subplot(131);plt.imshow(image[...,::-1]);plt.title("image")
plt.subplot(132);plt.imshow(added[...,::-1]);plt.title("add")
plt.subplot(133);plt.imshow(sub[...,::-1]);plt.title("subtract")
plt.show()
```

```python
image=cv2.imread(r"C:\\Users\Student\Desktop\pic02\006.jpg")
alpha=float(input('enter a number'))
beta=int(input('enter a number'))
new_image=cv2.convertScaleAbs(image,alpha=alpha,beta=beta)
plt.figure(figsize=[15,5])
plt.subplot(121);plt.imshow(image[...,::-1]);plt.title("image1")
plt.subplot(122);plt.imshow(new_image[...,::-1]);plt.title("image2")
plt.show()
```

```python
image=cv2.imread(r"C:\\Users\Student\Desktop\pic02\006.jpg")
new_image2=np.zeros(image.shape,image.dtype)
alpha=float(input('enter a number'))
beta=int(input('enter a number'))
for y in range(image.shape[0]):
    for x in range(image.shape[1]):
        for c in range(image.shape[2]):
            new_image2[y,x,c]=np.clip(alpha*image[y,x,c]+beta,0,255)
plt.figure(figsize=[15,5])
plt.subplot(121);plt.imshow(image[...,::-1]);plt.title("image1")
plt.subplot(122);plt.imshow(new_image2[...,::-1]);plt.title("image2")
plt.show()       
```

```python
gray_image=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
new_image=cv2.cvtColor(new_image,cv2.COLOR_BGR2GRAY)
new_image2=cv2.cvtColor(new_image2,cv2.COLOR_BGR2GRAY)
plt.figure(figsize=[15,5])
plt.subplot(131);plt.hist(gray_image.ravel(),256,[0,255]);plt.title("image")
plt.subplot(132);plt.hist(new_image.ravel(),256,[0,255]);plt.title("new_image")
plt.subplot(133);plt.hist(new_image2.ravel(),256,[0,255]);plt.title("new_image2")
plt.show()  
```

```python
gamma=0.4
img=cv2.imread(r"C:\\Users\Student\Desktop\pic02\000.jpg")
lookuptable=np.empty((1,256),np.uint8)
for i in range(256):
    lookuptable[0,i]=np.clip(pow(i/255,gamma)*255,0,255)
out=cv2.LUT(img,lookuptable)
plt.figure(figsize=[15,5])
plt.subplot(121);plt.imshow(img[...,::-1]);plt.title("image")
plt.subplot(122);plt.imshow(out[...,::-1]);plt.title("output")
plt.show()  
```

```python
gamma=0.4
img1=cv2.imread(r"C:\\Users\Student\Desktop\pic02\a.jpg")
img2=cv2.imread(r"C:\\Users\Student\Desktop\pic02\aaa.jpg")
output1=cv2.convertScaleAbs(img1,alpha=1.3,beta=40)
output2=cv2.convertScaleAbs(img2,alpha=1.3,beta=40)
lookuptable=np.empty((1,256),np.uint8)
for i in range(256):
    lookuptable[0,i]=np.clip(pow(i/255,gamma)*255,0,255)
output3=cv2.LUT(img1,lookuptable)
output4=cv2.LUT(img2,lookuptable)
plt.figure(figsize=[15,5])
plt.subplot(231);plt.imshow(img1[...,::-1]);plt.title("image")
plt.subplot(232);plt.imshow(output1[...,::-1]);plt.title("output")
plt.subplot(233);plt.imshow(output3[...,::-1]);plt.title("output")

plt.subplot(234);plt.imshow(img2[...,::-1]);plt.title("output")
plt.subplot(235);plt.imshow(output2[...,::-1]);plt.title("output")
plt.subplot(236);plt.imshow(output4[...,::-1]);plt.title("output")
plt.tight_layout()
plt.show()  
```
