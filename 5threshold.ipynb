```python
image=cv2.imread(r"C:\\Users\Student\Desktop\pic03\gradient.jpg",0)
ret1,new1=cv2.threshold(image,127,255,cv2.THRESH_BINARY)
ret2,new2=cv2.threshold(image,127,255,cv2.THRESH_BINARY_INV)
ret3,new3=cv2.threshold(image,127,255,cv2.THRESH_TRUNC)
ret4,new4=cv2.threshold(image,127,255,cv2.THRESH_TOZERO)
ret5,new5=cv2.threshold(image,127,255,cv2.THRESH_TOZERO_INV)
plt.figure(figsize=[10,5])
plt.subplot(1,6,1);plt.imshow(image,cmap="gray");plt.title("orginal");
plt.subplot(1,6,2);plt.imshow(new1,cmap="gray");plt.title("THRESH_BINARY");
plt.subplot(1,6,3);plt.imshow(new2,cmap="gray");plt.title("THRESH_BINARY_INV");
plt.subplot(1,6,4);plt.imshow(new3,cmap="gray");plt.title("THRESH_TRUNC");
plt.subplot(1,6,5);plt.imshow(new4,cmap="gray");plt.title("THRESH_TOZERO");
plt.subplot(1,6,6);plt.imshow(new5,cmap="gray");plt.title("THRESH_TOZERO_INV");
plt.tight_layout()
plt.show()
```
