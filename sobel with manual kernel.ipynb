```python
# Edges

import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('002.jpeg',0)

kernel = np.array([[0,0,0], 
                   [-1/2,0,1/2], 
                   [0,0,0]])
#1/2,-1/2

edge = cv2.filter2D(image, -1 , kernel)
edges = cv2.filter2D(image, cv2.CV_64F , kernel)

plt.figure(figsize=[12,7])
plt.subplot(141);plt.imshow(image, cmap='gray');plt.title("Original");
plt.subplot(142);plt.imshow(edge, cmap='gray');plt.title("Edge");
plt.subplot(143);plt.imshow(edges, cmap='gray');plt.title("Edges");

```
