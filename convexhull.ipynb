```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
import random
```

```python
random.seed=13
random_points=np.random.randint(100,400,20)
random_points=random_points.reshape(10,2)
image=np.zeros((512,512,3),np.uint8)
for i in random_points:
    cv2.circle(image,(i[0],i[1]),8,(0,255,0),-1)
plt.imshow(image[...,::-1])
plt.show()
```

```python
hull=cv2.convexHull(random_points)
cv2.drawContours(image,[hull],-1,(0,255,255),4)
plt.imshow(image[...,::-1])
plt.show()
```

```python
bgr_image=cv2.imread(r"C:\\Users\Student\Desktop\pic04\shapes.png")
```

```python
gray=cv2.cvtColor(bgr_image,cv2.COLOR_BGR2GRAY)
```

```python
ret,thresh=cv2.threshold(gray,127,255,cv2.THRESH_BINARY)
```

```python
contours,hirarchy=cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
```

```python
hull_list=[]
for i in range(len(contours)):
    hull=cv2.convexHull(contours[i])
    hull_list.append(hull)
cv2.drawContours(bgr_image,contours,-1,(0,0,255),7)
cv2.drawContours(bgr_image,hull_list,-1,(0,255,0),2)
plt.imshow(bgr_image[...,::-1])
plt.show()
```
