```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt
```

```python
image=cv2.imread(r"C:\\Users\Student\Desktop\pic04\shapes.png",0)
```

```python
ret,thresh=cv2.threshold(image,127,255,cv2.THRESH_BINARY)
```

```python
contours,hirarchy=cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
```

```python
for c in contours:
    print(c.shape)
```

```python
bgr_image=cv2.cvtColor(image,cv2.COLOR_GRAY2BGR)
cv2.polylines(bgr_image,[contours[-1]],True,(0,0,255),3)
plt.imshow(bgr_image[...,::-1])
```

```python
bgr_image=cv2.cvtColor(image,cv2.COLOR_GRAY2BGR)
cv2.polylines(bgr_image,contours,True,(0,0,255),3)
plt.imshow(bgr_image[...,::-1])
```

```python
image=cv2.imread(r"C:\\Users\Student\Desktop\pic04\rectangle.png",0)
ret,thresh=cv2.threshold(image,127,255,cv2.THRESH_BINARY)
contours,hirarchy=cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_NONE)
bgr_image=cv2.cvtColor(image,cv2.COLOR_GRAY2BGR)
for point in contours[0]:
    cv2.circle(bgr_image,(point[0][0],point[0][1]),2,(0,255,0),-1)
plt.imshow(bgr_image[...,::-1])
plt.title("CHAIN_APPROX_NONE")
plt.show()
```

```python
image=cv2.imread(r"C:\\Users\Student\Desktop\pic04\rectangle.png",0)
ret,thresh=cv2.threshold(image,127,255,cv2.THRESH_BINARY)
contours,hirarchy=cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
bgr_image=cv2.cvtColor(image,cv2.COLOR_GRAY2BGR)
for point in contours[0]:
    cv2.circle(bgr_image,(point[0][0],point[0][1]),2,(0,255,0),-1)
plt.imshow(bgr_image[...,::-1])
plt.title("CHAIN_APPROX_SIMPLE")
plt.show()
```
