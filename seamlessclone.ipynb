```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def combine(image1,image2):
    foreground=image1.copy()
    background=image2.copy()
    foreground_height=foreground.shape[0]
    foreground_width=foreground.shape[1]
    alpha=0.5
    blended=cv2.addWeighted(foreground,alpha,background[:foreground_height,:foreground_width,:],1-alpha,0,background)
    background[:foreground_height,:foreground_width,:]=blended
    return background
```

```python
src=cv2.imread(r"C:\\Users\Student\Desktop\pic07\airplane.jpg")
dst=cv2.imread(r"C:\\Users\Student\Desktop\pic07\f01.jpeg")
result=combine(src,dst)
plt.imshow(result[...,::-1])
plt.show()
```

```python
src=cv2.imread(r"C:\\Users\Student\Desktop\pic07\airplane.jpg")
plt.imshow(src[...,::-1])
plt.show()
```

```python
poly=np.array([[2,10],[20,10],[70,8],[85,20],[70,25],[40,30],[4,25]],np.int32)
cv2.polylines(src,[poly],1,(0,0,255))
src_mask=np.zeros(src.shape,src.dtype)
cv2.fillPoly(src_mask,[poly],(255,255,255))
plt.figure(figsize=[12,7])
plt.subplot(121);plt.imshow(src[...,::-1]);plt.title("image")
plt.subplot(122);plt.imshow(src_mask[...,::-1]);plt.title("mask")
plt.show()
```

```python
center=(100,25)
output1=cv2.seamlessClone(src,dst,src_mask,center,cv2.NORMAL_CLONE)
output2=cv2.seamlessClone(src,dst,src_mask,center,cv2.MIXED_CLONE)
plt.figure(figsize=[12,7])
plt.subplot(121);plt.imshow(output1[...,::-1]);plt.title("normal")
plt.subplot(122);plt.imshow(output2[...,::-1]);plt.title("mixed")
plt.show()
```
