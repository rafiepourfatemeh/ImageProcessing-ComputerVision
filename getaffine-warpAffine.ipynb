```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
image=plt.imread(r"C:\\Users\Student\Desktop\pic\triangle.jpg")
row,cols,ch=image.shape
plt.imshow(image)
plt.show()
```

```python
ps1=np.float32([[100,350],[100,90],[490,350]])
ps2=np.float32([[200,400],[500,300],[200,200]])
```

```python
m=cv2.getAffineTransform(ps1,ps2)
dst=cv2.warpAffine(image,m,(cols,row))
```

```python
plt.subplot(121);plt.imshow(image);plt.title("image")
plt.subplot(122);plt.imshow(dst);plt.title("Affine Transform")
plt.show()
```
