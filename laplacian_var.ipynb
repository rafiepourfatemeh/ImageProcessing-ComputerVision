```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image1=cv2.imread(r"C:\\Users\Student\Desktop\pic07\004.jpeg",0)
image2=cv2.imread(r"C:\\Users\Student\Desktop\pic07\0014.jpg",0)
```

```python
def func01(image):
    return cv2.Laplacian(image,cv2.CV_64F).var()
```

```python
metric_image1=func01(image1)
metric_image2=func01(image2)
```

```python
plt.figure(figsize=[12,7])
plt.subplot(121);plt.imshow(image1,cmap="gray");plt.title(metric_image1)
plt.subplot(122);plt.imshow(image2,cmap="gray");plt.title(metric_image2)
plt.show()
```

```python
cap=cv2.VideoCapture(0)
while True:
    ret,frame=cap.read()
    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    metric=func01(gray)
    color=(0,0,255)
    if metric>53:
        color=(0,255,0)
        
    cv2.putText(frame,str(metric),(30,30),cv2.FONT_HERSHEY_COMPLEX,1,color)
    cv2.imshow("webcam",frame)
    if cv2.waitKey(1)&0xFF==27:
        break
cap.release()
cv2.destroyAllWindows()
```
