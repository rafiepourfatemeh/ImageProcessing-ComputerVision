```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('002.jpeg')
```

```python
kernel1 = np.array([[0,0,0], 
                    [0,1,0], 
                    [0,0,0]])
```

```python
k1 = cv2.filter2D(image, -1,kernel1)

plt.figure(figsize=[12,7])
plt.subplot(121);plt.imshow(image[...,::-1]);plt.title("Original");
plt.subplot(122);plt.imshow(k1[...,::-1]);plt.title("shape after filter");
```

```python
kernel2 = 1/9*np.array([[1,1,1], 
                              [1,1,1], 
                              [1,1,1]])
```

```python
k1 = cv2.filter2D(image, -1,kernel2)

plt.figure(figsize=[12,7])
plt.subplot(121);plt.imshow(image[...,::-1]);plt.title("Original");
plt.subplot(122);plt.imshow(k1[...,::-1]);plt.title("shape after filter");
```

```python
kernel1 = np.array([[0,0,0], 
                    [0,2,0], 
                    [0,0,0]])
kernel2 = 1/9*np.array([[1,1,1], 
                              [1,1,1], 
                              [1,1,1]])
final_kernel = kernel1-kernel2
k3 = cv2.filter2D(image, -1,final_kernel)

plt.figure(figsize=[12,7])
plt.subplot(121);plt.imshow(image[...,::-1]);plt.title("Original");
plt.subplot(122);plt.imshow(k3[...,::-1]);plt.title("Sharpened");
print(final_kernel)
```
