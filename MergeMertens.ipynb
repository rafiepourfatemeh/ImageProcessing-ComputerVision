```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_fn=[r"C:\\Users\Student\Desktop\pic07\a1.png",r"C:\\Users\Student\Desktop\pic07\a2.png",r"C:\\Users\Student\Desktop\pic07\a3.png",r"C:\\Users\Student\Desktop\pic07\a4.png"]
img_list=[cv2.imread (fn) for fn in img_fn]
exposure_times=np.array([15,2.5,0.25,0.0333],dtype=np.float32)
plt.figure(figsize=[18,4])
for i , (image,exposure) in enumerate(zip(img_list,exposure_times)):
    plt.subplot(1,4,i+1);plt.imshow(image[...,::-1]);plt.title(exposure);plt.xticks([]);plt.yticks([]);
```

```python
merge_mertens=cv2.createMergeMertens()
res_mertens=merge_mertens.process(img_list)
res_mertens8bit=np.clip(res_mertens*255,0,255).astype("uint8")
plt.figure(figsize=[8,8])
plt.imshow(res_mertens8bit[...,::-1]);plt.title("Mertens")
plt.show()
```
