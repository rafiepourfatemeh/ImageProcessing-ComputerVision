```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image=cv2.imread(r"C:\\Users\Student\Desktop\pic\color.png")
```

```python
b,g,r=cv2.split(image)
```

```python
plt.figure(figsize=[20,5])
plt.subplot(141);plt.imshow(r,cmap="gray");plt.title("red channel")
plt.subplot(142);plt.imshow(g,cmap="gray");plt.title("green channel")
plt.subplot(143);plt.imshow(b,cmap="gray");plt.title("blue channel")
plt.subplot(144);plt.imshow(image[...,::-1]);plt.title("image")
```

```python
image=cv2.imread(r"C:\\Users\Student\Desktop\pic\b.png")
b,g,r=cv2.split(image)
plt.figure(figsize=[20,5])
plt.subplot(141);plt.imshow(r,cmap="gray");plt.title("red channel")
plt.subplot(142);plt.imshow(g,cmap="gray");plt.title("green channel")
plt.subplot(143);plt.imshow(b,cmap="gray");plt.title("blue channel")
plt.subplot(144);plt.imshow(image[...,::-1]);plt.title("image")
```

```python
merge=cv2.merge([b,g,r])
plt.figure(figsize=[20,5])
plt.subplot(151);plt.imshow(r,cmap="gray");plt.title("red channel")
plt.subplot(152);plt.imshow(g,cmap="gray");plt.title("green channel")
plt.subplot(153);plt.imshow(b,cmap="gray");plt.title("blue channel")
plt.subplot(154);plt.imshow(merge[...,::-1]);plt.title("merge")
plt.subplot(155);plt.imshow(image[...,::-1]);plt.title("image")
```

```python
image=cv2.imread(r"C:\\Users\Student\Desktop\pic\x.jpg")
```

```python
image=cv2.imread(r"C:\\Users\Student\Desktop\pic\x.jpg")
b,g,r=cv2.split(image)
b_modify=b.astype("int16")+50
b_new=np.minimum(255,b_modify)
b_modify=b_new.astype("uint8")
merge1=cv2.merge([b_modify,g,r])
plt.figure(figsize=[20,5])
plt.subplot(151);plt.imshow(r,cmap="gray");plt.title("red channel")
plt.subplot(152);plt.imshow(g,cmap="gray");plt.title("green channel")
plt.subplot(153);plt.imshow(b_modify,cmap="gray");plt.title("blue channel")
plt.subplot(154);plt.imshow(image[...,::-1]);plt.title("image")
plt.subplot(155);plt.imshow(merge1[...,::-1]);plt.title("merge")
```

```python
image=cv2.imread(r"C:\\Users\Student\Desktop\pic\x.jpg")
b,g,r=cv2.split(image)
b_modify=cv2.add(b,30)
merge1=cv2.merge([b_modify,g,r])
plt.figure(figsize=[20,5])
plt.subplot(151);plt.imshow(r,cmap="gray");plt.title("red channel")
plt.subplot(152);plt.imshow(g,cmap="gray");plt.title("green channel")
plt.subplot(153);plt.imshow(b_modify,cmap="gray");plt.title("blue channel")
plt.subplot(154);plt.imshow(image[...,::-1]);plt.title("image")
plt.subplot(155);plt.imshow(merge1[...,::-1]);plt.title("merge")
```
