```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image=cv2.imread(r"C:\Users\Student\Desktop\pic\004.jpg")
plt.imshow(image[...,::-1])
plt.show()
```

```python
point_A=np.float32([[10,60],[150,20],[80,0],[210,140]])
point_B=np.float32([[0,50],[140,10],[70,0],[200,110]])
```

```python
m=cv2.getPerspectiveTransform(point_A,point_B)
warp=cv2.warpPerspective(image,m,(300,300))
cv2.imshow("result",warp)
cv2.waitKey()
cv2.destroyAllWindows()
plt.figure(figsize=[10,15])
plt.subplot(121);plt.imshow(image[...,::-1]);plt.title("image")
plt.subplot(122);plt.imshow(warp[...,::-1]);plt.title("output")
plt.show()
```
