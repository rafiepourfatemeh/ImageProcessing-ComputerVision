```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread(r"C:\\Users\Student\Desktop\pic04\noisy.jpg",0)
```

```python
ret1,th1=cv2.threshold(img,127,255,cv2.THRESH_BINARY)
ret2,th2=cv2.threshold(img,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
blur=cv2.GaussianBlur(img,(5,5),0)
ret3,th3=cv2.threshold(blur,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
```

```python
image=[img,0,th1,
       img,0,th2,
       blur,0,th3]
titles=["orginal","histogram","global threshold 127",
       "orginal  noisy image","histogram noisy image"," otsu noisy image",
        "orginal  GaussianBlur","histogram GaussianBlur"," otsu GaussianBlur"]
plt.figure(figsize=[15,7])
for i in range(3):
    plt.subplot(3,3,i*3+1),plt.imshow(image[i*3],'gray')
    plt.title(titles[i*3]),plt.xticks([]),plt.yticks([])
    plt.subplot(3,3,i*3+2),plt.hist(image[i*3].ravel(),256)
    plt.title(titles[i*3+1]),plt.xticks([]),plt.yticks([])    
    plt.subplot(3,3,i*3+3),plt.imshow(image[i*3+2],'gray')
    plt.title(titles[i*3+2]),plt.xticks([]),plt.yticks([])
plt.show()
```

```python
orginal_img=cv2.imread(r"C:\\Users\Student\Desktop\pic04\sudoku.jpg",0)
noise_remove=cv2.medianBlur(orginal_img,5)
_,output1=cv2.threshold(orginal_img,127,255,cv2.THRESH_BINARY)
output2=cv2.adaptiveThreshold(orginal_img,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,2)
output3=cv2.adaptiveThreshold(orginal_img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,11,2)

_,output4=cv2.threshold(noise_remove,127,255,cv2.THRESH_BINARY)
output5=cv2.adaptiveThreshold(noise_remove,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,2)
output6=cv2.adaptiveThreshold(noise_remove,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,11,2)
```

```python
plt.figure(figsize=[15,7])
plt.subplot(331);plt.imshow(orginal_img, cmap="gray");plt.title("orginal_img")
plt.subplot(332);plt.imshow(noise_remove, cmap="gray");plt.title("noise_remove")
plt.subplot(334);plt.imshow(output1, cmap="gray");plt.title("output1-THRESH_BINARY")
plt.subplot(335);plt.imshow(output2, cmap="gray");plt.title("output2-ADAPTIVE_THRESH_MEAN_C")
plt.subplot(336);plt.imshow(output3, cmap="gray");plt.title("output3-ADAPTIVE_THRESH_GAUSSIAN_C")
plt.subplot(337);plt.imshow(output4, cmap="gray");plt.title("output4-THRESH_BINARY noise_remove")
plt.subplot(338);plt.imshow(output5, cmap="gray");plt.title("output5-ADAPTIVE_THRESH_MEAN_C noise_remove")
plt.subplot(339);plt.imshow(output6, cmap="gray");plt.title("output6-ADAPTIVE_THRESH_GAUSSIAN_C noise_remove")
plt.tight_layout()
plt.show()
```
